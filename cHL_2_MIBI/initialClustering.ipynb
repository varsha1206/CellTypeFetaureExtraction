```python
import umap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler


# Normalize the data (UMAP works best with scaled data)
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(df[df.columns])

# Apply UMAP to reduce dimensions to 2D
umap_model = umap.UMAP(n_components=2,random_state=42)
embedding = umap_model.fit_transform(features_scaled)

# Add UMAP components to DataFrame
df["UMAP1"] = embedding[:, 0]
df["UMAP2"] = embedding[:, 1]

# Plot UMAP result
plt.figure(figsize=(10, 6))
sns.scatterplot(x="UMAP1", y="UMAP2", data=df, alpha=0.7)
plt.title("UMAP Projection")
plt.show()
```

## GMM Clustering

```python
from sklearn.preprocessing import MinMaxScaler
# Standardize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture

# Fit GMM with 6 clusters (since we know cell types = 6)
gmm = GaussianMixture(n_components=6,covariance_type='full', random_state=42)
gmm.fit(X_scaled)

# Predict cluster assignments
labels = gmm.predict(X_scaled)
df["Cluster"] = labels  # Add cluster labels to the original dataframe
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
from umap import UMAP

# Reduce dimensions using UMAP
umap = UMAP(n_components=2, min_dist=0.5, random_state=42)
X_umap = umap.fit_transform(X_scaled)

# Scatter plot of clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=labels, palette="Set2", s=20)
plt.title("GMM Clustering Visualization using UMAP")
plt.xlabel("UMAP Component 1")
plt.ylabel("UMAP Component 2")
plt.legend(title="Cluster")
plt.show()
```
